```python
import pandas as pd

df = pd.read_csv("C:/Users/Platini AGOUANET/OneDrive/Desktop/UQO-Automne/Fouille et entreposage/Projet 2/INF5173-P2-Equipe15/data/raw/Retail_Transactions_Dataset.csv")  # ou sample
print(df.head())
print(df.shape)
print(df.columns)
```

```python

summary = []
for col in df.columns:
    nunique = df[col].nunique()
    null_rate = df[col].isna().mean()
    sample_vals = df[col].dropna().astype(str).head(3).tolist()
    summary.append((col, df[col].dtype, nunique, f"{null_rate:.2%}", sample_vals))

prof = pd.DataFrame(summary, columns=["column","dtype","nunique","null_rate","samples"]).sort_values("nunique")
# Heuristique dimension potentielle:
prof["dimension_candidate"] = (
    (prof["dtype"].astype(str).str.contains("object|category")) &
    (prof["nunique"].between(2, 200000))  # seuil large;
)
print(prof.head(30))
```
